```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#our data contains some string or character ,we need to get convert those into numeric data
le=LabelEncoder()
data_clean['sex']=le.fit_transform(data_clean['sex'])
data_test['sex']=le.fit_transform(data_test['sex'])
```

```python
#As we saw above ,there are some columns which has NULL values we need to fill those 
#generally we take average of column to fill null values using fillna function

data_clean=data_clean.fillna(data_clean['age'].mean())
data_test=data_test.fillna(data_test['age'].mean())
```

```python

input_cols=['pclass','sex','age','sibsp','parch','fare']
output_cols=['survived']
X=data_clean[input_cols]
Y=data_clean[output_cols]
print(X.shape,Y.shape)
```

```python
#Now we are done with preprocessing of our data 
#We shall move to our decision Tree functions
```

```python
#Entropy is the randomness of our system 
def entropy(col): 
    counts = np.unique(col,return_counts=True)
    
    N = float(col.shape[0])
    
    ent = 0.0
    for ix in counts[1]:
        p = ix / N
        ent += (-1.0*p*np.log2(p))
    
    return ent
```

```python
#For our decision tree ,we need to divide our data in two parts at each node
def divide_data(x_data, fkey, fval):
    
    x_left = pd.DataFrame([], columns=x_data.columns)
    x_right = pd.DataFrame([], columns=x_data.columns)
    
    for ix in range(x_data.shape[0]):
              
        val = x_data[fkey].iloc[ix]
        
        if val <  fval:
            x_left = x_left.append(x_data.iloc[ix])
        else:
            x_right = x_right.append(x_data.iloc[ix])
            
    return x_left,x_right
```

```python
# from current data,information gain is calculated for each column in the data 
def information_gain(x_data, fkey, fval):
    
    left,right = divide_data(x_data, fkey, fval)
    
    l = float(left.shape[0]/x_data.shape[0])
    r = float(right.shape[0]/x_data.shape[0])
    
    if left.shape[0] == 0 or right.shape[0] == 0: #it means we have reached the leaf node ,
                                                  #Hence we returned a large -ve number to avoid further splitting
        return -100000  
    
    return entropy(x_data.survived) - (l*entropy(left.survived) + r*entropy(right.survived))
```

```python
#to test our information gain function
for fx in data_clean.columns:
    print (fx)
    print (information_gain(data_clean, fx, data_clean[fx].mean()))
```

```python
class DecisionTree:
    
    def __init__(self, depth=0,max_depth=5):
        
        self.left = None
        self.right = None
        self.fkey = None  #fkey represents the name of the column
        self.fval = None #fval represents the threshold value for that column
        self.depth = depth
        self.max_depth = max_depth #Parameter max-depth is taken so as to avoid overfitting
        self.target = None
        
    def train(self, X_train):
        
        
        features = ["pclass","sex","age","sibsp","parch","fare"]        
        
        inf_gains = []
        
        for ix in features:
            inf_gains.append((information_gain(X_train, ix, data_clean[ix].mean())))
        
        self.fkey = features[np.argmax(inf_gains)]
        self.fval = X_train[self.fkey].mean()
        
        print("Making tree feature",self.fkey)
        
        data_left,data_right = divide_data(X_train,self.fkey, self.fval)
        
        data_left = data_left.reset_index(drop=True)
        data_right = data_right.reset_index(drop=True)
        
        
        #In case of a truly leaf node
        if data_left.shape[0] == 0 or data_right.shape[0] == 0:
            if X_train.survived.mean() > 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
    
        #Stop early if current depth > maxdepth
        if self.depth >= self.max_depth:
            if X_train.survived.mean() > 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
    
        #recursive case
        self.left = DecisionTree(depth=self.depth+1, max_depth=self.max_depth)
        self.left.train(data_left)
        
        self.right = DecisionTree(depth=self.depth+1, max_depth=self.max_depth)
        self.right.train(data_right)
        
        if X_train.survived.mean() > 0.5:
            self.target = "Survive"
        else:
            self.target = "Dead"
        return
    
    def predict(self, test): #to predict for an guven input
        if test[self.fkey] > self.fval:
            if self.right == None:
                return self.target
            return self.right.predict(test)
        else:
            if self.left == None:
                return self.target
            return self.left.predict(test)
```

```python
dt=DecisionTree()
dt.train(data_clean)
```



# We have trained our data
